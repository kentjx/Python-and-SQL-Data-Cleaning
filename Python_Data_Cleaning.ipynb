# Python Data Cleaning
In this Exercise, we will attempt to do Data Cleaning with Python. The objective and dataset is identical to SQLQuery4DataCleaning.sql, however we will use python to achieve the same results. 

## Objective 
1. Filling Null values of 'PropertyAddress' based on ParcelID. The PropertyAddress is essentially mapped to unique ParcelIDs, However, there are some null values in the 'PropertyAddress'. We will map it according to ParcelIDs.  

```python
df1 = df.copy()
```

```python
df['PropertyAddress'].isnull().sum()
```

```python
df1.dropna(axis = 0, subset = ['PropertyAddress'], inplace = True) #dropna in df1 PropertyAddress
```

```python
df1.info() #confirmed dropped. 
```

```python
di = pd.Series(df1.PropertyAddress.values,index=df1.ParcelID).to_dict() #using the dropped df1, we create a dictionary with no 
#NUll values

df.PropertyAddress = df.PropertyAddress.fillna(df.ParcelID.map(di)) #filling up the proper df with the dictionary we created 
```

```python
df['PropertyAddress'].isnull().sum()
```

## Objective 
1. Breaking out Address into Individual Columns (Address, City, State) 

```python
df.head()
```

```python
df['Town'] = df['PropertyAddress'].str.split(',').apply(lambda x: x[-1])
```

```python
df['Address'] = df['PropertyAddress'].str.split(',').apply(lambda x: x[0])
```

# Objective
1. Change Y and N to Yes and no in "Sold as Vacant" field

```python
df['SoldAsVacant'].value_counts()
```

```python
def function(x):
    if x == 'Y' or x == 'Yes':
        return 'Yes'
    else:
        return 'No'
    
df['SoldAsVacant'] =df['SoldAsVacant'].apply(function)
df['SoldAsVacant']
```

# Objective 
1. Remove Duplicates. The UniqueID is unique to all row entries. However, there are some entries that are obviously duplicates. Based on the condition that if parcelID, PropertyAddress, SalePrice,SaleDate,LegalReference. 

```python
df.drop_duplicates(subset = ['ParcelID', 'PropertyAddress', 'SalePrice', 'SaleDate', 'LegalReference'], inplace = True)
```

```python
df.info() #56373 entries. duplicates dropped. 
```
